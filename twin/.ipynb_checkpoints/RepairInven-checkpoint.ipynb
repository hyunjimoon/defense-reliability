```python
import numpy as np
import pandas as pd
import pysd #v.3.3.0
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
model = pysd.read_vensim('/Users/mdl')
```

변수명은 영어로 작성되어야 한다. 


```python
# running model
stocks = model.run()
stocks
```

```python
# final time 변경 가능, 원하는 variable의 결과값만 보기
# pysd는 변수명을 '소문자,_'로 자동변환한다. 
values = model.run(return_timestamps=range(200), return_columns = {'battle_field','inventory','repair_shop'})
values.plot(figsize=(8,8))
```

```python
# setting parameter values
# constant : 예방정비주기를 5 -> 10으로 변경
test1 = model.run(params = {'predictive_maintenance_period' : 10},
                  return_timestamps = range(200),
                  return_columns = {'battle_field', 'inventory', 'repair_shop'})
test1.plot(figsize=(8,8))
```

```python
# setting initial conditions ; stock variable의 초기값 변경 가능
test2 = model.run(initial_condition = (0, {'repair_shop' : 2}),
                  return_timestamps = range(200),
                  return_columns = {'battle_field', 'inventory', 'repair_shop'})
test2.plot(figsize=(8,8))
test2
```

```python
# querying current values
model.components.repair_shop()
```

```python
# replacing model components with more complex objects

# pysd는 모든 variable을 함수형태로 인식한다
stocks[['Maintenance Time']]

# 다음의 함수 형식으로 읽혀진다
# def maintenance_time():
#     return 1

# 그래서 함수 형식으로 바꿔주면 된다
def new_maintenance_time():
    return 2

model.components.maintenance_time = new_maintenance_time

test4 = model.run()
test4[['Maintenance Time']]
```

```python
# P-system

model2 = pysd.read_vensim('/Users/ksnam/Desktop/KNDU life/Paper/2. Paper/220620_maintenance inventory model/vensim model/repair model with inventory2.mdl')
stocks2 = model2.run()
stocks2
```

```python
# 재고일수 변화에 따른 battle field 변화

def stock_days(x) :
    stock = model2.run(params = {'n_of_days_in_stock' : x},
                       return_timestamps = range(300),
                       return_columns = {'battle_field'})
    x= stock.index.to_list()
    y= stock[['battle_field']]
    plt.plot(x,y)
    plt.legend(x)
    
fig = plt.figure(figsize=(8, 8))
list = [5,10,11,12,13,15,20]
for x in list:
    stock_days(x)
```
