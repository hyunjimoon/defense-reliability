```python
# battle field change according to inventory time

def stock_days(x) :
    stock = model2.run(params = {'n_of_days_in_stock' : x},
                       return_timestamps = range(300),
                       return_columns = {'battle_field'})
    x= stock.index.to_list()
    y= stock[['battle_field']]
    plt.plot(x,y)
    plt.legend(x)
    
fig = plt.figure(figsize=(8, 8))
list = [5,10,11,12,13,15,20]
for x in list:
    stock_days(x)
```
